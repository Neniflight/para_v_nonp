```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon
```

# Normal Distribution
## mean = 50, SD = 10

```python
normal_dataset = np.round(np.random.normal(loc=50, scale=10, size=100000), decimals=2)
np.random.shuffle(normal_dataset)
```

```python
x = np.linspace(min(normal_dataset), max(normal_dataset), 1000)
pdf = (norm.pdf(x, 50, 10))
plt.plot(x, pdf, color='r', label='True Distribution')

plt.hist(normal_dataset, bins=100, density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.grid(True)
plt.show()
plt.show()
```

# Exponential Distribution
## starts at 1, rate = 10

```python
exp_dataset = np.round(np.random.exponential(scale=10, size=100000) + 1, decimals=2)
np.random.shuffle(exp_dataset)

exp_dataset
```

```python
x = np.linspace(min(exp_dataset), max(exp_dataset), 1000)
pdf = (expon.pdf(x, 1, 10))
plt.plot(x, pdf, color='r', label='True Distribution')

plt.hist(exp_dataset, bins=100, density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Exponential Distribution')
plt.grid(True)
plt.show()
plt.show()
```

# Bimodal Distribution
## two peaks at 30 and 100
### created by two normal distribution of with mean of 30 and sd of 8 and mean of 100 and sd of 10

```python
mu1, sigma1 = 30, 8  
mu2, sigma2 = 100, 10 

samples1 = np.random.normal(mu1, sigma1, 90000)
samples2 = np.random.normal(mu2, sigma2, 10000)

combined_samples = np.round(np.concatenate((samples1, samples2)), decimals=2)
np.random.shuffle(combined_samples)


x = np.linspace(min(combined_samples), max(combined_samples), 1000)
pdf = (0.9 * norm.pdf(x, mu1, sigma1)) + (0.1 * norm.pdf(x, mu2, sigma2))
plt.plot(x, pdf, color='r', label='True Distribution')

plt.hist(combined_samples, bins=100, density=True)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Bimodal Distribution')
plt.grid(True)
plt.show()
```
